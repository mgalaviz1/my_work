ANSWER: For ENROLL, we can impute a mean by year except for the first and last year which are missing all values for ENROLL. We can backfill 1992 and forward fill 2017.  

The revenue columns and the expenditure columns are missing mostly for certain areas in the US so we should remove areas in the US that contain mostly missing values (value counts < 15 because imputing a value would introduce too much error. Remaining nulls in these columns can be imputed with the median since we're dealing with money. 

The average columns are missing every other year so we should interpolate the mean of the year before and after.
