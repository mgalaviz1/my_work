The average columns are missing every other year so we impute the mean for all years since we are missing so much data. We could also choose to drop these missing values and limit our analysis to every other year but I'll choose use the overall mean so that we can still analyze other columns.

Q3. This time, fill in the missing values using interpolation (extrapolation).

ANSWER:  We can interpolate for all missing values in the columns except that US territories or areas not part of the 50 US states should be removed because they are missing 64% of their values. For revenue and expenditures they're missing all values so it would be best to remove them. The remaining columns with missing values will be interpolated because the data is ordered by year or ordered by trends in time. 

```python
df.columns
```

```python
# interpolating enrollment column
df["ENROLL"] = df["ENROLL"].interpolate(method='time')
```

```python
df['ENROLL'].isnull().sum()
```

```python
# interpolating revenue columns
df["TOTAL_REVENUE"] = df["TOTAL_REVENUE"].interpolate(method='ffill')
df["FEDERAL_REVENUE"] = df["FEDERAL_REVENUE"].interpolate(method='ffill')
df["STATE_REVENUE"] = df["STATE_REVENUE"].interpolate(method='ffill')
df["LOCAL_REVENUE"] = df["LOCAL_REVENUE"].interpolate(method='ffill')
```

```python
df.isnull().sum()
```

```python
# interpolating expenditure columns
df["TOTAL_EXPENDITURE"] = df["TOTAL_EXPENDITURE"].interpolate(method='ffill')
df["INSTRUCTION_EXPENDITURE"] = df["INSTRUCTION_EXPENDITURE"].interpolate(method='ffill')
df["SUPPORT_SERVICES_EXPENDITURE"] = df["SUPPORT_SERVICES_EXPENDITURE"].interpolate(method='ffill')
df["OTHER_EXPENDITURE"] = df["OTHER_EXPENDITURE"].interpolate(method='ffill')
df["CAPITAL_OUTLAY_EXPENDITURE"] = df["CAPITAL_OUTLAY_EXPENDITURE"].interpolate(method='ffill')
```

```python
df.isnull().sum()
```

```python
df["OTHER_EXPENDITURE"] = df["OTHER_EXPENDITURE"].interpolate(method='bfill')
```

```python
# interpolating grades columns
df["GRADES_PK_G"] = df["GRADES_PK_G"].interpolate(method='ffill')
df["GRADES_KG_G"] = df["GRADES_KG_G"].interpolate(method='ffill')
df["GRADES_4_G"] = df["GRADES_4_G"].interpolate(method='ffill')
df["GRADES_8_G"] = df["GRADES_8_G"].interpolate(method='ffill')
df["GRADES_12_G"] = df["GRADES_12_G"].interpolate(method='ffill')
df["GRADES_1_8_G"] = df["GRADES_1_8_G"].interpolate(method='ffill')
df["GRADES_9_12_G"] = df["GRADES_9_12_G"].interpolate(method='ffill')
df["GRADES_ALL_G"] = df["GRADES_ALL_G"].interpolate(method='ffill')
```
